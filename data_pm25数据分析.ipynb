```python
import urllib.request
import re
import urllib.error
import time
import matplotlib.pyplot as plt
headers = headers=("User-Agent","Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/45.0.2454.101 Safari/537.36")
opener = urllib.request.build_opener()
opener.addheaders=[headers]
urllib.request.install_opener(opener)
'''def getpm25_plt(city):
    get_pm25 = '<div class="span1">\s{1,}<div class="value">\n\s{1,}(.*?)\s{1,}</div>'
    getpm25 = re.compile(get_pm25,re.S).findall(data)
    plt.plot(getpm25[0])
    plt.show()'''
def getpm25(city):
    url = 'http://pm25.in/'+ city
    data=urllib.request.urlopen(url).read().decode("utf-8")
    print(city)
    get_pm25 = '<div class="span1">\s{1,}<div class="value">\n\s{1,}(.*?)\s{1,}</div>'
    getpm25 = re.compile(get_pm25,re.S).findall(data)
    print('AQI指数，PM2.5，PM10，  CO，    NO2，  SO2')
    print(getpm25)
    data_time = '<div class="live_data_time">\s{1,}<p>数据更新时间：(.*?)</p>'
    datatime=re.compile(data_time,re.S).findall(data)
    print(datatime)    
    f = open('data_pm25.csv','a')
    f.write(datatime[0])
    f.write(',')
    for pm25 in getpm25:
        f.write(str(pm25))
        f.write(',')
    f.write('\n')
    f.close()
    return datatime
while True:
    print('开始采集')
    getpm25('beijing')
    time.sleep(3600)
```
